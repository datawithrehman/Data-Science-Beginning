# Imputing Missing Values in the Titanic Dataset (Beginner's Guide)


This notebook provides a simple, step-by-step guide for beginners on how to handle missing values in a dataset using the pandas library in Python. We will use a simplified version of the famous Titanic dataset to demonstrate common imputation techniques.

Missing data is a common problem in real-world datasets. It can lead to errors in your analysis or machine learning models if not handled properly. Imputation is the process of filling in these missing values.


## 1. Prepare Data and Environment


First, let's import the necessary libraries and create a sample DataFrame that resembles the Titanic dataset. In a real scenario, you would load your dataset from a CSV file (e.g., `pd.read_csv('titanic.csv')`) and you can also load through seaborn library (`sns.load_dataset('titanic')`).


```python

import pandas as pd
import numpy as np
import seaborn as sns

#Load Data Set through SEaborn library 

df = sns.load_dataset('titanic')

#Show the first 10 rows of the DataFrame

print("Original DataFrame (first 10 rows):")
df.head(10)
```

## 2. Identify Missing Values


Before we can impute, we need to know where the missing values are. We can use `isnull()` (or `isna()`) combined with `sum()` to count missing values per column.


```python

print("Missing values before imputation:")
print(df.isnull().sum())
```

## 3. Demonstrate Imputation Methods


Now, let's apply some common imputation techniques to our dataset.

### a. Imputing 'Age' with the Mean

The 'Age' column is numerical and has some missing values. A common strategy for numerical data is to fill missing values with the mean (average) of the existing values in that column.


```python

# Impute missing 'Age' values with the mean
df['age'].fillna(df['age'].mean(), inplace=True)

print("Missing values after Age imputation:")
print(df.isnull().sum())
```


### b. Imputing 'Embarked' with the Mode

The 'Embarked' column is categorical (S, C, Q) and has a few missing values. For categorical data, a good strategy is to fill missing values with the mode (the most frequent value) of the column.


```python

# Impute missing 'Embarked' values with the mode
# .mode()[0] is used because mode() can return multiple values if they have the same frequency
df['embarked'].fillna(df['embarked'].mode()[0], inplace=True)

print("Missing values after Embarked imputation:")
print(df.isnull().sum())
```


### c. Handling 'Deck' Column

The 'Deck' column has a significant number of missing values. When a column has too many missing values, imputing them might not be meaningful, and it's often better to drop the column entirely or consider it as a separate feature. For this beginner's guide, we will drop it.


```python

# Drop the 'Cabin' column due to too many missing values
df.drop("deck", axis=1, inplace=True)

print("Missing values after dropping deck column:")
print(df.isnull().sum())
```


### d. Final Check

Let's do a final check to ensure all missing values have been handled.


```python

print("Final check for missing values:")
print(df.isnull().sum())

print("Cleaned DataFrame (first 10 rows):")
df.head(10)
```
