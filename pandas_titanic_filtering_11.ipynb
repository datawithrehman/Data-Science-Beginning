# üìò Intermediate Filtering with Pandas (Titanic Dataset)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# ‚úÖ Load Dataset (Titanic)
df = sns.load_dataset('titanic')
df.head()
```

```python
# ‚úÖ Basic Overview
df.info()
```

```python
# ‚úÖ Filter Rows: Passengers older than 30
df[df['age'] > 30].head()
```

```python
# ‚úÖ Filter Rows: Females only
df[df['sex'] == 'female'].head()
```

```python
# ‚úÖ Filter Rows: Males under 18 (Boys)
df[(df['sex'] == 'male') & (df['Age'] < 18)].head()
```

```python
# ‚úÖ OR Condition: First Class OR Survived
# Using | for OR (remember to wrap each condition in parentheses)
df[(df['pclass'] == 1) | (df['Survived'] == 1)].head()
```

```python
# ‚úÖ Complex Condition: Females in 1st class who survived
df[(df['sex'] == 'female') & (df['pclass'] == 1) & (df['survived'] == 1)].head()
```

```python
# ‚úÖ Using .query(): Clean syntax filtering
df.query("sex == 'male' and age < 18 and survived == 1").head()
```

```python
# ‚úÖ Filter by multiple values: Pclass in [1, 2]
df[df['pclass'].isin([1, 2])].head()
```

```python
# ‚úÖ Missing Data Filter: Show rows with missing Age
df[df['age'].isnull()].head()
```

```python
# ‚úÖ Negation (~): Passengers *not* in 3rd class
df[~(df['pclass'] == 3)].head()
```

```python
# ‚úÖ Combine Filters Step-by-Step
adults = df[df['age'] >= 18]
adults_females = adults[adults['sex'] == 'female']
adults_females.head()
```

```python
# ‚úÖ Summary: Key Techniques
# - Boolean filtering with &, |, ~
# - .isin() for list matching
# - .str.contains() for text filtering
# - .isnull() to catch missing values
# - .query() for clean expressions
```
