# üìä Visualizing Data with Pandas (Using `tips` Dataset)

Welcome! In this notebook, we‚Äôll learn how to use **pandas** for quick and easy visualizations using the famous `tips` dataset from seaborn.

*You don't need to be a plotting expert ‚Äî just a few lines of code will do the job.*

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## üì• Load the Dataset

We'll use the built-in `tips` dataset, which contains information about restaurant tips.

```python
# Load tips dataset (requires internet access)
df = sns.load_dataset('tips')
df.head()
```

## üìà Line Plot

Let‚Äôs create a basic line plot of total bills.

```python
df['total_bill'].plot(title='Total Bill Trend', figsize=(10, 5))
plt.xlabel('Index')
plt.ylabel('Total Bill')
plt.grid(True)
plt.show()
```

## üìä Bar Plot

Now let‚Äôs see the average tip per day.

```python
df.groupby('day')['tip'].mean().plot(kind='bar', color='skyblue', title='Average Tip by Day')
plt.ylabel('Average Tip ($)')
plt.xlabel('Day')
plt.grid(axis='y')
plt.show()
```

## üìâ Histogram

Visualize the distribution of total bills using a histogram.

```python
df['total_bill'].plot(kind='hist', bins=20, color='salmon', title='Distribution of Total Bill')
plt.xlabel('Total Bill ($)')
plt.show()
```

## üíæ Save Your Plot

You can also save your plot as an image.

```python
plot = df.groupby('day')['total_bill'].sum().plot(kind='bar', color='orange', title='Total Sales by Day')
plt.ylabel('Total Bill ($)')
plt.xlabel('Day')
plt.tight_layout()
plt.savefig('total_sales_by_day.png')
plt.show()
```
